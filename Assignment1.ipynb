2.Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
lst=[]
for i in range(2000,3201): # iterate 2000 to 3200 in for loop
    if(i%7==0 and i%5 != 0):       #to check number is divisible by 7 and not a multiple of 5     
        lst.append(str(i))

print(','.join(lst) )          
                  
        
```

the the reverse order with a space between first name and last name

```python
firstName=input("enter firstname:") #To take first name
lastName=input("enter lastname:")   #To take second name
Name= firstName + ' ' + lastName   # add first and last name with space between them
print(Name[::-1])                  #prints string in reverse
```
